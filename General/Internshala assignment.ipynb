```python
import requests

r = requests.get("https://www.sec.gov/Archives/edgar/data/35315/0000035315-03-000001.txt")
print "page obtained"
from bs4 import BeautifulSoup

soup = BeautifulSoup(r.text,'lxml')

regtag=list()

for tag in soup.find_all('td'):
    for child in tag.childGenerator():
        try:
            list1=child.find_all('font')
            for i in list1 :
                if (str(i.text) == "Credit Default Swap"):
                    regtag.append(i.text)
        except Exception as e:
            print e
        

val=list()

for k in regtag:
    try:
        if(len(k.text)>3):
            val.append(k)    
        
    except Exception as e:
        pass
       #print e

    
print "values list obtained"
trtags=list()
fields=list()

for i in val:
    trtags.append(i.parent.parent.parent.parent)

print "trtags ready"
for each_tag in trtags:
    fields.append(each_tag.next_sibling.next_sibling)
print "final parsing"
for j in fields:
    for i in j.children :
        try:
            print i.p.font.get_text()
        except Exception as e:
            pass
           # print e
            
print "done"
        







```
