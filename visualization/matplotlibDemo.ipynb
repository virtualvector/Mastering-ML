```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
```

```python
y=x**2
```

```python
plt.plot(x,y,'r--')
plt.xlabel('xlabel')
plt.ylabel('ylabel')
plt.title('title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r-')
plt.subplot(1,2,2)
plt.plot(y,x)
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)


axes.set_xlabel('hello')
axes.set_ylabel('ylabel')
axes.set_title('title')
plt.show

```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.3,0.3])
axes1.plot(x,y,'r-')
axes2.plot(y,x)
```

```python

fig,axes = plt.subplots(nrows=2,ncols=1)
axes[0].plot(y,x,'r-')
axes[1].plot(x,y)
```

```python
fig = plt.figure(figsize=(6,4))
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.6,0.2,0.2]) 
axes3 = fig.add_axes([0.6,0.2,0.2,0.2])
axes.plot(x,y,label='x and y')
axes.plot(y,x,'r',label = 'y and x')
axes2.plot(y,x,'r--')
axes2.plot(x,y)
axes3.plot(x,y)
axes3.plot(y,x,'r--')
axes.legend()
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y,color='#006600',linewidth=0.7
          ,ls='-',marker='+',markersize=20)

```
